```python
cols = ['Sales', 'Profit', 'Discounted_price']  # define column

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower, upper)
```

```python
sns.boxplot(x=df['Discounted_price'])
```

- insight : Outlier treatment using IQR improved data stability, reduced variance, and made the pricing distribution more consistent. However, care must be taken as capping may hide extreme but valuable business cases such as premium transactions

# Future Product Demand Prediction

```python
# Category-wise sales in each month
data = df.groupby(["Month", "Category"])["Sales"].sum().reset_index()

# Total sales of each month
monthly_total = df.groupby("Month")["Sales"].sum().reset_index()
monthly_total.columns = ["Month", "Total_Monthly_Sales"]

# Merge total sales
data = pd.merge(data, monthly_total, on="Month")

# Calculate percentage based on total monthly sales
data["Sales_Percentage"] = (
    data["Sales"] / data["Total_Monthly_Sales"]
) * 100

# Arrange month order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

data["Month"] = pd.Categorical(
    data["Month"],
    categories=month_order,
    ordered=True
)

# Sort values
data = data.sort_values(
    ["Month", "Sales_Percentage"],
    ascending=[True, False]
)

# Final Output
print(data[[
    "Month",
    "Category",
    "Sales",
    "Sales_Percentage"
]])
```
